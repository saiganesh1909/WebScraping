```python
- The project is to scrap the data using Library BeautifulSoup and requests 

- We are scraping the IMDB website

Step-1: Import the all the Libraries

from bs4 import BeautifulSoup
import requests
import lxml

- have to take the url on which website we to scrape the data 

url = "https://www.imdb.com/chart/top/"

- We have to take the data from url by using "Requests"

data = requests.get(url)

- Just see the data which is stored on the data variable

print(data.text) # here "text" we have to use because we have to fetch the all the data

- We get the all the data + code but we have to only fetch the data what we requried.

- this is done by using BeautifulSoup and parse the data with lxml

- we have to contruct the Beautifulsoup 

soup = BeautifulSoup(data.text,'lxml')

print(soup)

- We have to  fetch the data and store into the excel file first we havr to fetch

movies = soup.find('tbody',class_='lister-list').find_all('tr')

movies

- 1. Rank number
- 2. Title of the movie
- 3. Rating
-4.  Year

for i in movies:
    number=i.find('td',class_="titleColumn").get_text(strip=True).split('.')[0]
    title=i.find('td',class_="titleColumn").find('a').text
    year = i.find('td',class_="titleColumn").find('span').text.strip('( )')
    rating = i.find('td',class_="ratingColumn imdbRating").text.strip(' ')
    sheet.append([number,title,year,rating])
    

- We fetch the data what we requried so now we have to load the data to excel file y importing "openpyxl" library

import openpyxl

- We have to create the workbook by using the library

excel = openpyxl.Workbook() 

print(excel.sheetnames)

sheet =excel.active
sheet.title="Imdb Movie Ratings"
sheet.append(['Movie Rank','Movie Name','Year of Realse','Rating'])

excel.save('Movie Ratings.xlsx')
```
